Matplotlib

*   useful for making plots

```python
import matplotlib.pyplot as plt
```

```python
# to get the data we import numpy
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
print(x)
print(y)
print(z)
```

**plotting the data**

```python
#plotting the sin wave
plt.plot(x,y)
plt.show()
```

```python
#plotting the cos wave
plt.plot(x,z)
plt.show()
```

```python
# to add the title for x and y axis
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sin value')
plt.title('sine wave')
plt.show()
```

```python
#plotting a parabola, it will print 20 equal spaced values in an array starting from -10 to 10
x = np.linspace(-10, 10, 20)
y = x**2
plt.plot(x,y)
plt.show
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
plt.plot(x,y,'g--')
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

**Bar plotting**

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1] )  # 0,0 -> ,coordinates, is the origin and 1,1 -> is the hieght and width of the rect
languages = ['python', 'java', 'c++', 'ruby', 'javascript']
students = [23,17,35,29,12]
ax.bar(languages, students)
plt.xlabel('languages')
plt.ylabel('number of students')
plt.title('bar plot')
plt.show()
```

**Pie chart**

```python
fig1 = plt.figure()
ax1 = fig1.add_axes([0,0,1,1])
languages = ['python', 'java', 'c++', 'ruby', 'javascript']
students = [23,17,35,29,12]
ax1.pie(students, labels = languages, autopct='%1.2f%%') #2f - floating binary
plt.show()
```

**Scatter plot**

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax2 = fig2.add_axes([0,0,1,1])
ax2.scatter(x,y,color='g')
ax2.scatter(x,z,color='b')
plt.show()
```

**3D Scatter plot**

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata)
ydata = np.cos(zdata)
ax.scatter3D(xdata, ydata, zdata,c=zdata,cmap='Blues')
plt.show()
```
