**Seaborn**

*   used for data visualization

```python
 import seaborn as sns
 import matplotlib.pyplot as plt
 import numpy as numpy
 import pandas as pd
```

seaborn has few built in datasets

```python
#total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.set_theme()
#gives a better theme, visual
```

```python
#visualizing thos data
sns.relplot(data = tips, x = 'total_bill', y = 'tip',col ='time',hue = 'smoker',style = 'smoker',size ='size')
#compared to matplotlib, seaborn automatically categorizes things, we dont need to specify
```

```python
#loading the iris dataet
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

scatter plot

```python
sns.scatterplot(x='sepal_length',y='petal_width',hue='species',data=iris)
```

```python
#loading the titanic dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

count plot

```python
sns.countplot(x='class',data=titanic)
```

bar chart

```python
sns.barplot(x='sex',y='survived',hue = 'class',data=titanic)
```

```python
#house price data set
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()

house = pd.DataFrame(housing.data,columns=housing.feature_names)
house['price'] = housing.target
house.head()
```

```python
print(housing)
```

```python
house.head()
```

distribution plot

```python
sns.displot(house['price'])
```

correlation -

*  postitve(if one value increases the othe value also increases)

*   negative()

  

```python
correlation = house.corr()
```

heat map -  +ve correlation

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
#light colour means negative corerelation and dark colours mean positive corelation
```
